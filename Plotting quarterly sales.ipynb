```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
q = ['Q1', 'Q2', 'Q3', 'Q4']
y = {2017: 0, 2018: 1, 2019: 2}
sales = np.array([
    [425, 366, 498, 403],
    [486, 385, 546, 432],
    [474, 390, 538, 412]
])

plt.rcParams['figure.figsize'] = (12,6)
X = np.arange(1, 5)
plt.grid(ls='--')
plt.bar(X - 0.25, sales[y[2017]], width=0.2)
plt.bar(X,        sales[y[2018]], width=0.2, tick_label=q)
plt.bar(X + 0.25, sales[y[2019]], width=0.2)

plt.legend(loc=(0.85,0.8), labels=y.keys())
plt.show()
```

```python
plt.grid(ls='--')
plt.bar(y.values(), sales[:, 0], tick_label=list(y.keys()))
plt.bar(y.values(), sales[:, 1], bottom=sales[:, 0])
plt.bar(y.values(), sales[:, 2], bottom=sales[:, 0] + sales[:, 1])
plt.bar(y.values(), sales[:, 3], bottom=sales[:, 0] + sales[:, 1] + sales[:, 2])

plt.legend(loc=(0.98, 0.8), labels=q)
plt.show()
```
